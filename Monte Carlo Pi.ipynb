```python
import matplotlib.pyplot as plt
import numpy
import random
from decimal import *

getcontext().prec=28

hits = 0
throws = 0

print("Reset")
```

```python
x_points_hit=[]
y_points_hit=[]
x_points_miss=[]
y_points_miss=[]

plt.gcf().set_size_inches([5,5])

plt.axis([0,1,0,1])
for i in range(0,1000000):
#    x=random.SystemRandom().random()
#    y=random.SystemRandom().random()
    x=random.random()
    y=random.random()
    throws += 1
    if (x*x+y*y<=1):
        hits += 1
        if (i%1000==0):
            x_points_hit.append(x)
            y_points_hit.append(y)
    else:
        if (i%1000==0):
            x_points_miss.append(x)
            y_points_miss.append(y)

plt.scatter(x_points_hit, y_points_hit, marker=".")
plt.scatter(x_points_miss, y_points_miss, marker=".")
plt.show()

dhits = Decimal(hits)
dthrows = Decimal(throws)

dpi = dhits/dthrows*Decimal(4)

print("Hits:", hits)
print("Throws:", throws)
print("Pi Estimate", dpi)
```
